```python
import requests
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')


# Function to fetch book information from Google Books API
def fetch_book_info(title):
    url = f'https://www.googleapis.com/books/v1/volumes?q=intitle:{title}'
    response = requests.get(url)
    data = response.json()
    if 'items' in data and len(data['items']) > 0:
        return data['items'][0]['volumeInfo']
    else:
        return None

# Function to preprocess text data
def preprocess_text(text):
    if isinstance(text, str):  # Check if the text is a string
        # Tokenization
        tokens = word_tokenize(text)

        # Lowercasing
        tokens = [token.lower() for token in tokens]

        # Removing punctuation
        tokens = [token for token in tokens if token not in string.punctuation]

        # Removing stopwords
        stop_words = set(stopwords.words('english'))
        tokens = [token for token in tokens if token not in stop_words]

        # Lemmatization
        lemmatizer = WordNetLemmatizer()
        tokens = [lemmatizer.lemmatize(token) for token in tokens]

        # Join tokens back into text
        preprocessed_text = ' '.join(tokens)

        return preprocessed_text
    else:
        return ''  # Return an empty string if the input is not a string or is NaN


# Fetch book information from Google Books API
fetched_book_info = fetch_book_info('1984')

# Extract relevant features of the fetched book
fetched_title = fetched_book_info.get('title', '')
fetched_authors = fetched_book_info.get('authors', [])
fetched_description = fetched_book_info.get('description', '')

# Preprocess textual features of the fetched book
fetched_description = preprocess_text(fetched_description)
fetched_authors_str = ', '.join(fetched_authors)

# Combine all textual features into one string
fetched_text = f"{fetched_title} {fetched_authors_str} {fetched_description}"

```

## Transformer based recommendation engine

```python
import requests
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string
from sentence_transformers import SentenceTransformer
import torch

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Load pre-trained sentence transformer model
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Function to fetch book information from Google Books API
def fetch_book_info(title):
    print(f"Fetching information for book: {title}")
    url = f'https://www.googleapis.com/books/v1/volumes?q=intitle:{title}'
    response = requests.get(url)
    data = response.json()
    if 'items' in data and len(data['items']) > 0:
        return data['items'][0]['volumeInfo']
    else:
        print(f"No information found for book: {title}")
        return None

# Function to preprocess text data
def preprocess_text(text):
    if isinstance(text, str):  # Check if the text is a string
        # Tokenization
        tokens = word_tokenize(text)

        # Lowercasing
        tokens = [token.lower() for token in tokens]

        # Removing punctuation
        tokens = [token for token in tokens if token not in string.punctuation]

        # Removing stopwords
        stop_words = set(stopwords.words('english'))
        tokens = [token for token in tokens if token not in stop_words]

        # Lemmatization
        lemmatizer = WordNetLemmatizer()
        tokens = [lemmatizer.lemmatize(token) for token in tokens]

        # Join tokens back into text
        preprocessed_text = ' '.join(tokens)

        return preprocessed_text
    else:
        return ''  # Return an empty string if the input is not a string or is NaN

# List of books to fetch and compare
book_titles = ['1984', 'To Kill a Mockingbird', 'The Great Gatsby', 'Moby Dick', 'Pride and Prejudice',
               'The Catcher in the Rye', 'The Hobbit', 'Fahrenheit 451', 'Jane Eyre', 'Brave New World']

# Create a DataFrame to store the results
df_sim_transformer = pd.DataFrame(columns=['original_title', 'original_authors', 'original_description', 'similar_title', 'similar_authors', 'description', 'similarity_score'])

# Assuming you already have a DataFrame named book_df
# Preprocess textual features in the book_df DataFrame
print("Preprocessing book descriptions in the DataFrame")
book_df['preprocessed_description'] = book_df['description'].apply(preprocess_text)

# Loop over each book title in the list
for title in book_titles:
    fetched_book_info = fetch_book_info(title)

    if fetched_book_info:
        fetched_title = fetched_book_info.get('title', '')
        fetched_authors = fetched_book_info.get('authors', [])
        fetched_description = fetched_book_info.get('description', '')
        fetched_genre = fetched_book_info.get('categories', '')
        fetched_description = preprocess_text(fetched_description)
        fetched_authors_str = ', '.join(fetched_authors)
        fetched_text = f"{fetched_title} {fetched_authors_str} {fetched_description}"

        fetched_embedding = model.encode([fetched_text])

        similarity_scores = []

        for idx, row in book_df.iterrows():
            description = row['preprocessed_description']
            description_embedding = model.encode([description])
            similarity_score = torch.cosine_similarity(torch.tensor(fetched_embedding), torch.tensor(description_embedding)).item()
            similarity_scores.append(similarity_score)

        top_indices = sorted(range(len(similarity_scores)), key=lambda i: similarity_scores[i], reverse=True)[:10]

        similar_books_list = []
        for idx in top_indices:
            similar_books_list.append({
                'original_title': fetched_title,
                'original_authors': fetched_authors_str,
                'original_description': fetched_description,
                'original_genre': fetched_genre,
                'similar_title': book_df.iloc[idx]['title'],
                'similar_authors': book_df.iloc[idx]['authors'],
                'similar_genre':book_df.iloc[idx]['categories'],
                'description': book_df.iloc[idx]['description'],
                'similarity_score': similarity_scores[idx]
            })

        similar_books_df = pd.DataFrame(similar_books_list)
        df_sim_transformer = pd.concat([df_sim_transformer, similar_books_df], ignore_index=True)

        print(f"Similar books for '{title}' added to the DataFrame:")
        print(similar_books_df)

print("Final Results DataFrame:")
print(df_sim_transformer)
```

```python
#This column is a list but all the lists contain only 1 item so we turn it into a string
df_sim_transformer['original_genre'] = df_sim_transformer['original_genre'].apply(lambda x: x[0] if isinstance(x, list) else x)
```

Extracting keywords from descriptions to look for common keywords

```python
import re
import string
import nltk
import numpy as np
from nltk.corpus import stopwords
from transformers import pipeline

# Download NLTK resources
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Initialize keyword extractor pipeline
keyword_extractor = pipeline('feature-extraction', model='distilbert-base-uncased', tokenizer='distilbert-base-uncased')

def extract_keywords(text, top_n=5):
    if not isinstance(text, str) or not text:  # Check if the text is None, NaN, or empty
        return ''

    # Preprocess text
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = text.translate(str.maketrans('', '', string.punctuation.replace("'", "").replace("“", "").replace("”", "")))  # Remove punctuation except ' and special quotes
    text = text.replace("'", " ' ")  # Add space around single quotes
    text = re.sub(r'\s+', ' ', text)  # Remove extra whitespace

    # Tokenize text
    tokens = text.split()

    # Remove stopwords and special quotes
    tokens = [token for token in tokens if token not in stop_words and token != "'"]

    # Reconstruct preprocessed text
    preprocessed_text = ' '.join(tokens)

    # Extract features for the preprocessed text
    features = keyword_extractor(preprocessed_text)

    # Get the token embeddings
    embeddings = features[0]

    # Calculate the average embedding across all tokens
    avg_embedding = np.mean(embeddings, axis=0)

    # Calculate token importance by dot product of token embedding and average embedding
    token_importance = [(token, np.dot(embedding, avg_embedding)) for token, embedding in zip(preprocessed_text.split(), embeddings)]

    # Sort tokens by importance
    token_importance.sort(key=lambda x: x[1], reverse=True)

    # Return the top N tokens as keywords
    keywords = [token for token, _ in token_importance[:top_n] if token not in {'book','every','title' ,'theme','\'s','novel', 'winner','copy','year','prize'}]
    return ' '.join(keywords)

# Apply the keyword extraction to each description
df_sim_transformer['original_keywords'] = df_sim_transformer['original_description'].apply(lambda x: extract_keywords(x, top_n=10))
df_sim_transformer['keywords'] = df_sim_transformer['description'].apply(lambda x: extract_keywords(x, top_n=10))

# Display the dataframe with keywords
print(df_sim_transformer[['description', 'keywords']])
```

```python
df_sim_transformer[['description', 'keywords']]
```

```python
# Function to count common keywords
def count_common_keywords(row):
    # Split strings into lists of keywords
    original_keywords = row['original_keywords'].split()
    keywords = row['keywords'].split()

    # Convert lists into sets
    original_keywords_set = set(original_keywords)
    keywords_set = set(keywords)

    # Calculate intersection of sets
    common_keywords = original_keywords_set.intersection(keywords_set)

    # Count the number of common keywords
    num_common_keywords = len(common_keywords)

    return num_common_keywords

# Apply the function to each row and create a new column 'num_common_keywords'
df_sim_transformer['num_common_keywords'] = df_sim_transformer.apply(count_common_keywords, axis=1)
```

```python
# Check if there are common keywords, authors, or genres
has_common_keywords = df_sim_transformer['num_common_keywords'] > 0
has_common_authors = df_sim_transformer['original_authors'] == df_sim_transformer['similar_authors']
has_common_genre = df_sim_transformer['original_genre'] == df_sim_transformer['similar_genre']
similarity = df_sim_transformer['similarity_score']>=0.5

# Combine the conditions using logical OR operator
df_sim_transformer['relevance'] = has_common_keywords | has_common_authors | has_common_genre | similarity

# Convert boolean values to True/False strings
df_sim_transformer['relevance'] = df_sim_transformer['relevance'].map({True: 'True', False: 'False'})
```

## Cosine similarity based recommendation engine

```python
import requests
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Function to fetch book information from Google Books API
def fetch_book_info(title):
    print(f"Fetching information for book: {title}")
    url = f'https://www.googleapis.com/books/v1/volumes?q=intitle:{title}'
    response = requests.get(url)
    data = response.json()
    if 'items' in data and len(data['items']) > 0:
        return data['items'][0]['volumeInfo']
    else:
        print(f"No information found for book: {title}")
        return None

# Function to preprocess text data
def preprocess_text(text):
    if isinstance(text, str):  # Check if the text is a string
        # Tokenization
        tokens = word_tokenize(text)

        # Lowercasing
        tokens = [token.lower() for token in tokens]

        # Removing punctuation
        tokens = [token for token in tokens if token not in string.punctuation]

        # Removing stopwords
        stop_words = set(stopwords.words('english'))
        tokens = [token for token in tokens if token not in stop_words]

        # Lemmatization
        lemmatizer = WordNetLemmatizer()
        tokens = [lemmatizer.lemmatize(token) for token in tokens]

        # Join tokens back into text
        preprocessed_text = ' '.join(tokens)

        return preprocessed_text
    else:
        return ''  # Return an empty string if the input is not a string or is NaN

# List of books to fetch and compare
book_titles = ['1984', 'To Kill a Mockingbird', 'The Great Gatsby', 'Moby Dick', 'Pride and Prejudice',
               'The Catcher in the Rye', 'The Hobbit', 'Fahrenheit 451', 'Jane Eyre', 'Brave New World']

# Create a DataFrame to store the results
df_sim_cos = pd.DataFrame(columns=['original_title', 'original_authors', 'original_description', 'similar_title', 'similar_authors', 'description', 'similarity_score', 'original_genre', 'similar_genre'])

# Assuming you already have a DataFrame named book_df
# Preprocess textual features in the book_df DataFrame
print("Preprocessing book descriptions in the DataFrame")
book_df['preprocessed_description'] = book_df['description'].apply(preprocess_text)
book_df['preprocessed_authors'] = book_df['authors'].apply(preprocess_text)

# Combine textual features of books in the DataFrame into one string
book_df['text'] = book_df['title'] + ' ' + book_df['preprocessed_authors'] + ' ' + book_df['preprocessed_description']
book_df['text'] = book_df['text'].fillna('')

# Loop over each book title in the list
for title in book_titles:
    fetched_book_info = fetch_book_info(title)

    if fetched_book_info:
        fetched_title = fetched_book_info.get('title', '')
        fetched_authors = fetched_book_info.get('authors', [])
        fetched_description = fetched_book_info.get('description', '')
        fetched_genre = fetched_book_info.get('categories', [])

        fetched_description = preprocess_text(fetched_description)
        fetched_authors_str = ', '.join(fetched_authors)
        fetched_genre_str = ', '.join(fetched_genre)

        fetched_text = f"{fetched_title} {fetched_authors_str} {fetched_description}"

        # Calculate TF-IDF matrix for the fetched book
        tfidf_vectorizer = TfidfVectorizer()
        tfidf_matrix = tfidf_vectorizer.fit_transform(book_df['text'])
        fetched_tfidf_matrix = tfidf_vectorizer.transform([fetched_text])

        # Calculate cosine similarity between the fetched book and books in the DataFrame
        cosine_sim = cosine_similarity(fetched_tfidf_matrix, tfidf_matrix)

        # Get indices of similar books
        similar_books_indices = cosine_sim.argsort()[0][-10:][::-1]

        similar_books_list = []
        for idx in similar_books_indices:
            similar_books_list.append({
                'original_title': fetched_title,
                'original_authors': fetched_authors_str,
                'original_description': fetched_description,
                'original_genre': fetched_genre_str,
                'similar_title': book_df.iloc[idx]['title'],
                'similar_authors': book_df.iloc[idx]['authors'],
                'similar_genre': book_df.iloc[idx]['categories'],
                'description': book_df.iloc[idx]['description'],
                'similarity_score': cosine_sim[0][idx]
            })

        similar_books_df = pd.DataFrame(similar_books_list)
        df_sim_cos = pd.concat([df_sim_cos, similar_books_df], ignore_index=True)

        print(f"Similar books for '{title}' added to the DataFrame:")
        print(similar_books_df)

print("Final Results DataFrame:")
print(df_sim_cos)
```
